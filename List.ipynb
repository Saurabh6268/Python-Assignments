```python
#34 Count vowels in a string.

str_1="OpenAI"
vowels=['a','e','i','o','u','A','E','I','O','U']
count=0
for ch in str_1:
  if ch in vowels:
    count+=1
print("Number of vowels:",count)
```

```python
#40 Compare sum of digits in two strings

str_1="123"
str_2="321"
sum1 = sum(int(x) for x in str_1)
sum2 = sum(int(x) for x in str_2)

if sum1 == sum2:
  print("Equal")
else:
  print("Not equal")
```
